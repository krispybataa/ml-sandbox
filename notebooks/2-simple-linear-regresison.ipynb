# 2 | Simple Linear Regression
This notebook will detail a simple linear regression script run on a sample dataset. 

Simple Linear Regerssion is used to predict a response using a single feature 

## 2.1 Data Preprocessing

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

df = pd.read_csv('datasets/2-studentscores.csv')
X = df.iloc[ :, : 1].values
Y = df.iloc[ :, 1].values

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=1/4, random_state=0)
```

## 2.2 Fitting Model to training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor = regressor.fit(X_train,Y_train)
```

## 2.3 Result Prediction

```python
Y_pred = regressor.predict(X_test)
```

## 2.4 Visualization

### 2.4.1 Visualizing Training Results 

```python
plt.scatter(X_train, Y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
```

### 2.4.2 Visualizing Test Results

```python
plt.scatter(X_test, Y_test, color = 'red')
plt.plot(X_test, regressor.predict(X_test), color = 'blue')
```
